```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
```

```python
x0 = np.array([2,4,5,7])*0.1
w = np.array([1,2,-4,1])

x1 = np.array([ 0, 0, 1, 0, 2,-4, 0, 1, 0])
x2 = np.array([ 0, 0, 1, 1, 3,-1,-1, 0, 0])
x3 = np.array([ 0, 0, 0, 1, 2, 5, 4, 3, 3])

x = np.arange(0,1.1,0.125)
```

```python


ax = plt.figure(figsize=[6,3])
plt.stem(x,x1,basefmt=" ")

plt.xlim([0,1])
plt.ylim([-5,6])
plt.xlabel('Time t')
plt.ylabel('a.u.')
plt.grid(which='both')
plt.legend([r'$w_{inc}$'], loc='upper left')
plt.xticks(x)

plt.savefig('timeencode_min_example_1.png',dpi=600, bbox_inches = "tight")
plt.savefig('timeencode_min_example_1.pdf',format='pdf', bbox_inches = "tight")
```

```python

ax = plt.figure(figsize=[6,3])
plt.stem(x,x1,basefmt=" ")
plt.step(x,x2,'g', where='post')
plt.xlim([0,1])
plt.ylim([-5,6])
plt.xlabel('Time t')
plt.ylabel('a.u.')
plt.grid(which='both')
plt.legend([r'$\Delta w$',r'$w_{inc}$'], loc='upper left')
plt.xticks(x)

plt.savefig('timeencode_min_example_2.png',dpi=600, bbox_inches = "tight")
plt.savefig('timeencode_min_example_2.pdf',format='pdf', bbox_inches = "tight")
```

```python
ax = plt.figure(figsize=[6,3])
plt.stem(x,x1,basefmt=" ")
plt.step(x,x2,'g', where='post')
plt.plot(x,x3,'r')
plt.xlim([0,1])
plt.ylim([-5,6])
plt.xlabel('Time t')
plt.ylabel('a.u.')
plt.grid(which='both')
plt.legend([r'$\Delta w$',r'$u$',r'$w_{inc}$'], loc='upper left')
plt.xticks(x)

plt.savefig('timeencode_min_example_3.png',dpi=600, bbox_inches = "tight")
plt.savefig('timeencode_min_example_3.pdf',format='pdf', bbox_inches = "tight")
```
